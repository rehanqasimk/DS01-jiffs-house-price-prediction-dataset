```python
#cleaning the data
data2['land_size_sqm'] = data2['land_size_sqm'].apply(lambda x:x.replace(' sqm','')).astype('float')
data2['house_size_sqm'] = data2['house_size_sqm'].apply(lambda x:x.replace(' sqm','')).astype('float')
```

```python
def ynconvert(x):
    if x=='Yes':
        return 1
    else:
        return 0
```

```python
#loop for converting all the yes no in to 1 and 0
for i in data2.columns:
    if data2[i].unique().shape[0] == 2:
        data2[i] = data2[i].apply(ynconvert).astype('int')
```

```python
data2['distance_to_school'] = data2['distance_to_school'].apply(lambda x:x.replace('km','')).astype('float')
data2['distance_to_supermarket'] = data2['distance_to_supermarket'].apply(lambda x:x.replace(' km','')).astype('float')
data2['crime_rate_index'] = data2['crime_rate_index'].astype('float')
data2['house_age'] = data2['house_age'].apply(lambda x:x.replace(' Years','')).astype('int64')
```

```python
def roomsize(x):
    if x=='small':
        return 0
    elif x=='medium':
        return 1
    elif x=='large':
        return 2
    else:
        return 3
data2['room_size'] = data2['room_size'].apply(roomsize).astype('int')
```
